```python
import pandas as pd

df = pd.read_csv("test_Arabic_tweets", encoding="utf-8", header=None,names=["sentiment", "text"])
```

```python
import emoji
```

```python
def clean_text(text):

    text = emoji.replace_emoji(text,replace="")
    text = re.sub(r"http\s+|www.\s+", "" , text)
    text = re.sub(r"[^\w\s]","", text)
    text = re.sub(r"\d+", "" , text)
    text = re.sub(r"\s+", " " , text).strip()

    return text

df["clean_text"]= df["text"].apply(clean_text)
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline
```

```python
model_name = "CAMeL-Lab/bert-base-arabic-camelbert-da-sentiment" # "asafaya/bert-base-arabic"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
```

```python
sentiment_pipline = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
```

```python
texts = df['clean_text'].tolist()
```

```python
results = sentiment_pipline(texts, truncation=True)
```

```python
df['predicted_sentiment'] = [res['label'] for res in results]
```

```python
df[['clean_text', 'predicted_sentiment']].head(10)
```
